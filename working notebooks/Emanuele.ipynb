```python
def transform_query(firm, ngo, separator):
    firm = firm.split('/')[0]
    ngo = ngo.split('/')[0]
    string = f'{firm} {ngo}'
    string = ''.join(e for e in string if e.isalnum() or e.isspace())
    string = string.lower()
    query = string.split()
    query = separator.join(query)
    return query

def transform_date(date):
    date = str(date)
    formats = [
        '%Y-%m-%d',  # YYYY-MM-DD
        '%Y/%m/%d',  # YYYY/MM/DD
        '%d-%m-%Y',  # DD-MM-YYYY
        '%d/%m/%Y',  # DD/MM/YYYY
        '%Y-%m-%d %H:%M',  # YYYY-MM-DD HH:MM
        '%Y/%m/%d %H:%M',  # YYYY/MM/DD HH:MM
        '%Y-%m-%d %H:%M:%S',  # YYYY-MM-DD HH:MM:SS
        '%Y/%m/%d %H:%M:%S',  # YYYY/MM/DD HH:MM:SS
        '%d-%m-%Y %H:%M:%S',  # DD-MM-YYYY HH:MM:SS
        '%d/%m/%Y %H:%M:%S',  # DD/MM/YYYY HH:MM:SS
    ]

    for fmt in formats:
        try:
            date = datetime.strptime(date, fmt)
            break  # Exit the loop if parsing is successful
        except ValueError:
            pass  # Continue to the next format if parsing fails

    # If none of the formats matched, raise an error
    else:
        raise ValueError(f"Invalid date format: {date}")
    year = date.year
    month = date.month
    day = date.day
    return year, month, day

def next_date(date):

    year, month, day = transform_date(date)

    date = datetime(year, month, day)
    future_date = date + timedelta(days=1)

    future_year = future_date.year
    future_month = future_date.month
    future_day = future_date.day

    return future_year, future_month, future_day

    # Format the date as yyyy-mm-dd
    formatted_date = specific_date.strftime("%Y-%m-%d")
    print("Formatted specific date:", formatted_date)
```

```python
def scrape_financial_times(firm, ngo, date):
    driver = webdriver.Chrome(options=chrome_options)

    query = transform_query(firm, ngo, '+')

    year, month, day = transform_date(date)
    future_year, future_month, future_day = next_date(date)

    url = f'https://www.ft.com/search?sort=relevance&q={query}&dateTo={future_year}-{future_month}-{future_day}&dateFrom={year}-{month}-{day}'
    driver.get(url)
    soup = BeautifulSoup(driver.page_source, 'html.parser')
    articles = []
    list_articles = []
    try:
        pagination_text = soup.find('span', class_='search-pagination__page').get_text()
        number_pages = int(pagination_text.split('of')[-1].strip())
    except:
        number_pages = 1
    for i in range(1, number_pages+1):
        driver.get(f'https://www.ft.com/search?sort=relevance&q={firm}+{ngo}&dateTo={year}-{month}-{day}&dateFrom={year}-{month}-{day}&page={i}&sort=relevance&isFirstView=true')
        soup = BeautifulSoup(driver.page_source, 'html.parser')
        articles += soup.find_all('li', class_='search-results__list-item')
    for article in articles:
            heading = article.find('a', class_='js-teaser-heading-link').get_text().replace('\n', '').replace('\t', '')
            try:
                content = article.find('a', class_='js-teaser-standfirst-link').get_text().replace('\n', '').replace('\t', '')
            except:
                content = None
            dictionary = {'heading': heading,'content': content}

            list_articles.append(dictionary)

    return pd.DataFrame(list_articles)
```

```python
def scrape_figaro(firm, ngo, date):
    driver = webdriver.Chrome(options=chrome_options)

    query = transform_query(firm, ngo, '%20')

    year, month, day = transform_date(date)
    future_year, future_month, future_day = next_date(date)

    url = f'https://recherche.lefigaro.fr/recherche/{query}/?datemin={day}-{month}-{year}&datemax={future_day}-{future_month}-{future_year}'
    driver.get(url)
    soup = BeautifulSoup(driver.page_source, 'html.parser')
    articles = []
    list_articles = []
    articles += soup.find_all('article', class_='fig-profil fig-profil-mtpd')
    for article in articles:
        heading = article.find('h2', class_='fig-profil-headline').get_text().replace('\n', '').replace('\t', '')
        try:
            content = article.find('div', class_='fig-profil-chapo').get_text().replace('\n', '').replace('\t', '')
        except:
            content = None
        dictionary = {'heading': heading,'content': content}

        list_articles.append(dictionary)

    return pd.DataFrame(list_articles)

scrape_figaro('greenpeace', 'vert', '2024-10-11')
```

```python
def scrape_repubblica(firm, ngo, date):
    driver = webdriver.Chrome(options=chrome_options)

    query = transform_query(firm, ngo, '+')

    year, month, day = transform_date(date)
    future_year, future_month, future_day = next_date(date)

    url = f'https://ricerca.repubblica.it/ricerca/repubblica?query={query}&fromdate={year}-{month}-{day}&todate={future_year}-{future_month}-{future_day}&sortby=score&author=&mode=any'
    driver.get(url)
    soup = BeautifulSoup(driver.page_source, 'html.parser')
    articles = []
    list_articles = []
    articles += soup.find_all('article')
    for article in articles:
        heading = article.find('a').get_text().replace('\n', '').replace('\t', '')
        try:
            content = article.find('p').get_text().replace('\n', '').replace('\t', '')
        except:
            content = None
        dictionary = {'heading': heading,'content': content}

        list_articles.append(dictionary)

    return pd.DataFrame(list_articles)

scrape_repubblica('greenpeace', 'attack', '2024-11-04')
```

```python
def total_scrape(firm, ngo, date):
    df1 = scrape_financial_times(firm, ngo, date)
    df2 = scrape_figaro(firm, ngo, date)
    df3 = scrape_repubblica(firm, ngo, date)
    #print(df3)
    #print(len(df3))
    print(len(df1), len(df2), len(df3))
    return len(df1), len(df2), len(df3)
```
